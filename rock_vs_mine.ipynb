**Import** **libraries**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

**Load** **dataset**

```python
sonar=pd.read_csv('/content/sonar data.csv')
sonar.head(5)
```

**Data** **preprocessing**

```python
sonar.shape
```

```python
le=LabelEncoder()
sonar['R']=le.fit_transform(sonar['R'])
sonar.head(5)
```

```python
x= sonar.drop(columns='R',axis=1)
y=sonar['R']
print(x)
print(y)
```

**Train**-**Test** **Split**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

**Model** **Training**

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

**Model** **Evaluation**

```python
xtrain_acc=model.predict(x_train)
training_data_accuracy=accuracy_score(xtrain_acc,y_train)
print('accuracy score of training data:',training_data_accuracy)
```

```python
xtest_acc=model.predict(x_test)
testing_data_accuracy=accuracy_score(xtest_acc,y_test)
print('accuracy score of testing data:',testing_data_accuracy)
```
